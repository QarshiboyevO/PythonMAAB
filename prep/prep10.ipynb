```python
class Date:
    def __init__(self,year,month,day):
        try:
            if not(day<1 or day>31):
                if not(month<1 or month>12):
                    self.year=year
                    self.month=month
                    self.day=day
                else:
                    raise TypeError("Invalid date")
            else:
                raise TypeError("Invalid date")
        except TypeError:
            raise TypeError("Invalid date")
    def format1(self):
        return f"{self.day}/{self.month}/{self.year}"

    def format2(self):
        months=["January",'February','March','April','May','June','July','August','September','October','November','December']
        for i in range(len(months)):
            if i==self.month-1:
                return f"{months[i]} {self.day}, {self.year}"
    def format3(self):
        months=["January",'February','March','April','May','June','July','August','September','October','November','December']
        for i in range(len(months)):
            if i==self.month-1:
                return f"{self.day} {months[i]} {self.year}"

a=Date(2021,1,12)
print(a.format1())
print(a.format2())
print(a.format3())
```
