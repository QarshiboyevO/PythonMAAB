```python
import numpy as np
from matplotlib.style.core import library


def converttocelcius(farenheit):
    return (farenheit - 32)*5/9
vecconvertion=np.vectorize(converttocelcius)
farenheit=[32,68,100,212,77]
print(vecconvertion(farenheit))
```

```python
def powers(a,b):
    return a**b

vecpower=np.vectorize(powers)
a=[2,3,4,5]
b=[1,2,3,4]
print(vecpower(a,b))
```
