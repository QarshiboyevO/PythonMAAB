```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
s1 = Student("Ali",18)
s2 = Student("Bob",19)
print(s1.name, s1.age, s2.name, s2.age)
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hi, my name is {self.name}", f"\n I am {self.age} years old.")
s1 = Student("Baurjan", 19)
s1.greet()
```
